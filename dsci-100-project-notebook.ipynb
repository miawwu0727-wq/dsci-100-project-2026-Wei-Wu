#### Task 5.2 Perform the Analysis (20 points)

- Perform the proposed analysis.
- Be sure to comment your code throughout so that it is easy to follow. 

*You may add more code cells if needed.* 


</div>

```r
set.seed(68280734) #ensure the reproducibility

madness_split <- initial_split(tidy_madness, prop = 0.75, strata = champ) #split the dataset into train (75%) and test set (25%)
madness_train <- training(madness_split)
madness_test  <- testing(madness_split)

madness_folds <- vfold_cv(madness_train, v = 5, strata = champ) #create 5 cross-validation folds from train data to evaluate different values of K (to find best) 

#start build a workflow, build recipe to predict the 'champ', predictors is 'reg_wins_pct' and 'seed' as predictors
madness_recipe <- recipe(champ ~ reg_wins_pct + seed, data = madness_train) |>
  step_scale(all_numeric_predictors()) |> 
  step_center(all_numeric_predictors()) #standardize all numeric predictors 

#keep build workflow, set knn model specification (kknn/classifcation etc.)
knn_spec <- nearest_neighbor(weight_func = "rectangular", neighbors = tune()) |> #test a grid of possible K values to find the optimal k by tune()
  set_engine("kknn") |>
  set_mode("classification")

#keep build workflow, bundle recipe and knn model specification together to a single workflow
madness_workflow <- workflow() |>
  add_recipe(madness_recipe) |>
  add_model(knn_spec)

k_grid <- grid_regular(neighbors(range = c(1, 50)), levels = 10) #create a grid, test which k performs best

# tune the workflow by testing it across the grid of K values using 5 cross-validation folds, evaluating the most accurate k
knn_tune <- tune_grid(
  madness_workflow,
  resamples = madness_folds,
  grid = k_grid)

best_k <- select_best(knn_tune, metric = "accuracy")
final_workflow <- finalize_workflow(madness_workflow, best_k) #update our initial workflow with this optimal k
final_fit <- last_fit(final_workflow, split = madness_split) #perform the final fit, evaluating performance on test set.

collect_metrics(final_fit) #print the overall result
```
